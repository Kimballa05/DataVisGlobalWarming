```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
import plotly.io as pio
pio.renderers.default = 'notebook'
plt.style.use('ggplot')
%matplotlib inline
```

Making a radial bubble plot using literacy data and life expectancy. Bubble size is life expectancy and radius is literacy rate scale.

```python
data = pd.read_csv('Other Data - Copy.csv')

data
```

```python
data = data.dropna()

data
```

```python
# List of countries to remove
countries_to_remove = ['Australia', 'Central African Republic', 'Eswatini']

# Remove rows based on country names
data = data[~data['Country name'].isin(countries_to_remove)]
```

```python
# Selection box tool
selection = alt.selection_interval(encodings=['x', 'y'])

# Scatter plot
scatter = alt.Chart(data).mark_circle().encode(
    x=alt.X('Literacy rate (2011):Q', scale=alt.Scale(zero=False)),  # Remove extra space on x-axis
    y=alt.Y('Life expectancy (2011):Q', scale=alt.Scale(zero=False)),
    tooltip=['Country name', 'Literacy rate (2011)', 'Life expectancy (2011)'],
    color='Continent:N',  # Specify the column and its data type as Nominal (N)
    opacity=alt.condition(selection, alt.value(1), alt.value(0.5))  # Adjust opacity for selected vs. unselected points
).add_params(selection)



# Bar chart based on selected points from the scatter plot, color-coded by Continent
bar_chart = alt.Chart(data).transform_filter(selection).mark_bar().encode(
    x=alt.X('Country name', axis=alt.Axis(title=None)),
    y='Literacy rate (2011):Q',
    color='Continent:N'  # Specify the color encoding based on Continent
).properties(width=400)

# Second Bar chart based on selected points from the scatter plot, color-coded by Continent
bar_chart_2 = alt.Chart(data).transform_filter(selection).mark_bar().encode(
    x=alt.X('Country name', axis=alt.Axis(title=None)),
    y='Democracy Index (2011):Q',
    color='Continent:N'  # Specify the color encoding based on Continent
).properties(width=400)

# Third Bar chart based on selected points from the scatter plot, color-coded by Continent
bar_chart_3 = alt.Chart(data).transform_filter(selection).mark_bar().encode(
    x=alt.X('Country name', axis=alt.Axis(title=None)),
    y='GDP per capita (2011):Q',
    color='Continent:N'  # Specify the color encoding based on Continent
).properties(width=400)


# Display both scatter plot and bar chart
top_row = alt.hconcat(scatter, bar_chart)
bottom_row = alt.hconcat(bar_chart_2, bar_chart_3)
alt.vconcat(top_row, bottom_row)

```
